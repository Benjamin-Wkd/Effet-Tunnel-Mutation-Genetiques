# Modélisation du problème

Les mutations de l'ADN auxquelles on attribue un caractère aléatoire est en fait régie par l'effet tunnel en physique quantique[1]. En effet, les couples de nuclétodides A-T et C-G sont liés par liaisons hydrogènes et forment la double hélice d'ADN. Or pour une même liaison A-T, il existe deux positions stables pour le proton à l'origine de cette liaison hydrogène (Annexe-1). Graphiquement, elles correspondent aux puits de potentiel de la figure 1. Ces deux puits sont séparés par une barrière de potentiel alors franchissable par effet tunnel. Connaître le comportement des particules dans un environnement donnée et donc leur fonction d'onde associée nécesssite de résoudre l'équation de Schrodinger.

## Modélisation

### Obention de l'équation

### Géométrie du domaine

On étudiera dans un premier temps le comportement d'un proton à l'intérieur du potentiel $V> \epsilon$ de largeur k. On étudie le système selon une seule dimension d'espace, soit $\Omega= ]0,k[ \subset \mathbb{R} $  l'espace dans lequel évolue la particule. La fonction d'onde est donc définie elle aussi dans $\Omega$.

Les paramètres dépendant de cette géométrie sont donc : 

$\bullet$La fonction d'onde : $\Phi(x)$ 

$\bullet$Le potentiel auquel est soumis le proton : $V(x)$

Le code (Annexe-2) génère un exemple d'allure du potentiel sur le domaine $\Omega$ dans le cas particulier d'un potentiel symétrique (double puits de même profondeur).

Remarque : En pratique, l'énergie potentielle d'un proton pour la liasion hydrogène entre A et T n'est pas symétrique. Une des deux formes tautomères est moins stable que l'autre (Annexe2). 

### Conditions aux bords

On se restreint à l'étude de la fonction d'onde à l'intérieur du segment $]0,k[$, c'est-à-dire à l'intérieur de la barrière de potentiel. Par continuité de la fonction d'onde aux deux extremités, on obtient les conditions aux bords suivantes : 

\begin{equation}
\begin{cases}
\phi(0)=A\e^{ik_10}+R\e^{-ik_10} \\
\phi(k)=B\e^{ik_2k}
\end{cases}
\end{equation}

Ces deux équations d'ondes correspondent respectivement aux ondes incidente + réfléchie ($k_1$ domaine $]-\infty,0[$) et transmise ($k_2$ domaine $]k,+\infty[$). 
Plus simplement, nous avons :

\begin{equation}
\begin{cases}
\phi(0)=\alpha \\
\phi(k)=\beta
\end{cases}
\end{equation}



# Mise en équations

## Adimensionnement

On effectue le changement de variable suivants :
\begin{equation}
x=kx^*
\end{equation}
De sorte que l'équation se réécrit : $$\frac{-h^2}{2mk^2}\frac{\partial^2 \phi(x)}{\partial {x^*}^2}+(V(x)-\epsilon)\phi(x)=0
$$ 

Le problème final que nous allons étudier est donc le suivant:

\begin{equation}
\begin{cases}
-\frac{\partial^2 \phi(x)}{\partial {x^*}^2}+(V(x)-\epsilon)\frac{2ma^2}{h^2}\phi(x)=0 \ \  sur \ \  \Omega^*= ]0,1[ \\
\phi(0)=\alpha,  \ \ \ \phi(1)=\beta  \\
\end{cases}
\end{equation}

## Relèvement

<span style="text-decoration: underline;">Remarque </span>: L'hypothèse considérée tout au lond de l'étude est que : $\forall x \in \Omega, \ V(x) > \epsilon$ 

On pose $\phi_0(x)=\alpha+(\beta -\alpha)x$ et on considère $\phi(x)=\phi_0(x)+\tilde \phi(x)$. 
Dans ce cas, $\tilde \phi$ est solution de l'équation et les conditions aux limites sont :
\begin{equation}
\begin{cases}
\tilde\phi(0)=0 \\
\tilde\phi(1)=0
\end{cases}
\end{equation}




## Formulation variationnelle

On note $$C(x)=(V(x)-\epsilon)\frac{2mk^2}{h^2} > 0\ \  car \ \ V(x)>\epsilon$$
On pose alors

\begin{array}{ccccc}
A & : & H^1_0(0,1)\times H^1_0(0,1) & \to \mathbb{R} \\
 & & (u,v) & \mapsto & \int_{\Omega} u'v'+ Cuv\\
\end{array}

et \begin{array}{ccccc}
L & : & H^1_0(0,1) & \to \mathbb{R} \\
 & & v & \mapsto & -\int_{\Omega} Cu_0  \\
\end{array}

On obtient alors la formulation variationnelle suivante:

(FV) Trouver $u \in H^1_0(0,1)(\Omega)$ tel que  $\forall v \in H^1_0(0,1)(\Omega)$, $A(u,v)= L(v)$

## Discrétisation du domaine

On choisit de discrétiser le segment $]0,1[$ selon un maillage uniforme en J+1 intervalles. Notre pas vaut alors $h=\frac{1}{J+1}$.  

# Simulations numériques

## Cas général

On veut approcher la solution d'un problème d'équations aux dérivées
partielles, dont on sait qu'il est bien posé grâce à l'analyse, à l'aide d'une
méthode d'éléments finies. 

La démarche est la suivante (les étapes sont numérotées Ci pour étape i du code) : 
- C0 : on installe et charge les modules nécessaires ;
- C1 : on définit les constantes ;
- C2 : on définit la géométrie ; 
- C3 : On fabrique le maillage ; 
- C4 : On exprime l'espace de résolution ; 
- C5 : on prend en compte les conditions au bord ;
- C6 : on implémente la formulation variationnelle ;
- C7 : on résout ; 
- C8 : on trace la solution.

<p style="color:red;">     C0 : Chargement des modules </p>

Listes des modules nécessaires à la résolution et l'exploitation des résultats :

- dolfin
- numpy
- fenics
- matplotlib

```python
from dolfin import *
from fenics import *
import numpy as np
import matplotlib.pyplot as plt
```

<p style="color:red;"> C1 : Définition des constantes </p>

On fixe les constantes : la longueur du segment $k$, l'énergie portée par la particule $\epsilon$ qu'on note E, sa masse $m$, la constante de Planck réduite $\hbar$ qu'on note hbar. 

```python
k=1E-10 #distance H--O de la liaison hydrogène.
V0=1E-19
E=0.985*V0  #Atttention, L'Energie doit être inférieure à la valeur minimale de la barrière sur 0,1.
m=1.67E-27  #proton
hbar=6.62607E-34/(2*np.pi)
a_1,b_1,c_1=-0.04,0.04,0.99
V0=1E-19
alpha=0.5
beta=0.5
u0= Expression('alpha+(beta-alpha)*x[0]', degree=2, alpha=alpha, beta=beta)
V = Expression('V0*(a_1*pow(x[0]-0.5,2)+b_1*(x[0]-0.5)+c_1)',degree=2,a_1=a_1,b_1=b_1,c_1=c_1,V0=V0)
#V = Expression('V0',degree=2,V0=V0) #Potentiel constant
C = Expression('(V-E)*2*m*pow(k,2)/pow(hbar,2)',degree=2,V=V,E=E,m=m,k=k,hbar=hbar)
f = Expression('-C*u0',degree=2,C=C,u0=u0)
```

<p style="color:red;"> C2 : Définition de la géométrie </p>  

Il s'agit d'un segment de longueur adimensionnée 1.

```python
X=1
```

<p style="color:red;"> C3 : Fabrication du maillage </p>

```python
nx=100
#nx = int(input("Taille du maillage (entier naturel):"))
assert type(nx)==int

mesh = UnitIntervalMesh(nx)

print("Tracé d'un maillage uniforme")
plt.figure(1)
plot(mesh, title="Maillage du segment")
```

<p style="color:red;"> C4 : Définition de l'espace Hh dans lequel on va chercher la solution </p>

```python
Hh = FunctionSpace(mesh, 'P', 1)
```

<p style="color:red;"> C5 : Prise en compte des conditions au bord </p>

```python
 def DirichletBoundary(x, on_boundary):
         tol = 1e-14
         return on_boundary and abs(x[0]) < tol
     # Enforcing u = u0 at x = 0
bc = DirichletBC(Hh, Constant(0), DirichletBoundary)
```

<p style="color:red;"> C6 : Codage de la formulation variationnelle </p>

```python
u = TrialFunction(Hh)
v = TestFunction(Hh)
L = (f*v)*dx
A = (dot(grad(u),grad(v))+(C*u*v))*dx
```

<p style="color:red;"> C7 : Résolution de l'approximation variationnelle (assemblage de la matrice de rigidité et résolution du système linéaire) </p>

```python
uh = Function(Hh)
solve(A == L, uh, bc)
U=uh+u0 #relèvement
```

<p style="color:red;"> C8 : Tracé de la solution </p>

```python
plt.figure(3)
plot(uh,title='Solution numérique uh')
plt.xlabel('x')
plt.figure(4)
plot(U,title='Solution numérique finale après relèvement')
plt.xlabel('x')


fig, ax1 = plt.subplots()

color = 'tab:green'
ax1.set_xlabel('distance x')
ax1.set_ylabel('Densité de probabilité (ua)', color=color)
plot(U**2,title='Densité de probabilité en fonction de la localisation du proton' ,color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  

color = 'tab:gray'
ax2.set_ylabel('potentiel', color=color)  
ax2.plot(x,z, color=color)
ax2.plot(x,len(x)*[E], color='red')
ax2.tick_params(axis='y', labelcolor=color)
plt.legend(['V(x)','E'])

fig.tight_layout()  
```

## Prise en compte de l'aléatoire

La probabilité pour la particule de franchir cette barrière de potentiel, c'est-à-dire de sauter d'un état à l'autre varie en  fonction de l'écart énergétique entre la barrière et l'energie du proton. Ici pour mesurer cet écart on choisit de prendre la valeur du potentiel au sommet, soit $\Delta E=max(V)-E$ . Etudions alors la variation de la densité de probabilité en fonction de l'Energie de la particule.

```python
#Génération alétoire de la valeur d'Energie.
#L'Energie est de l'ordre de grandeur de l'eV.
Delta=V0*0.01*np.random.random_sample() #Delta doit être nécessairement positif.
Delta2=V0*0.01*np.random.random_sample()
print(Delta, Delta2) 
```

```python
C1 = Expression('(Delta)*2*m*pow(k,2)/pow(hbar,2)',degree=2,Delta=Delta,m=m,k=k,hbar=hbar)
f1 = Expression('-C1*u0',degree=2,C1=C1,u0=u0)

C2 = Expression('(Delta2)*2*m*pow(k,2)/pow(hbar,2)',degree=2,Delta2=Delta2,m=m,k=k,hbar=hbar)
f2 = Expression('-C2*u0',degree=2,C2=C2,u0=u0)
```

<p style="color:red;"> Codage de la nouvelle formulation variationnelle </p>

```python
u1 = TrialFunction(Hh)
v1 = TestFunction(Hh)

L1 = (f1*v1)*dx
A1 = (dot(grad(u1),grad(v1))+(C1*u1*v1))*dx

L2 = (f2*v1)*dx
A2 = (dot(grad(u1),grad(v1))+(C2*u1*v1))*dx
```

```python
uh1 = Function(Hh)
solve(A1 == L1, uh1, bc)
U1=uh1+u0 #relèvement

uh2 = Function(Hh)
solve(A2 == L2, uh2, bc)
U2=uh2+u0 #relèvement
```

# Description des simulations

Le modèle "3.1 Cas général" met en lumière l'influence de la forme de la barrière de potentiel sur la densité de probabiltité de présence du proton. Pour une même énergie E du proton, la densité de probabilité est tracé pour le potentiel étudié **(figure-2)**. On la compare au cas d'un potentiel constant **(figure-1)**.

   Dans le cas d'un potentiel constant, la densité de probabilité décroît exponentiellement au fur et à mesure qu'on pénètre dans la barrière. 
    
   On voit néanmoins que pour la figure 2, proche de l'origine, il y a une croissance de la densité de probabilité due à la hauteur relative V(x)-E très faible. Puis quand cette hauteur relative augmente, le comportement s'assimile à celui à celui d'un potentiel constant, le proton ne pouvant plus distinguer les variations du potentiel.
    
   Néanmoins, comme l'illustre la **figure 3** cet effet tunnel n'est possible que si la barrière de potentiel n'est pas trop large. Auquel cas, le proton ne peut la franchir.

Le deuxième modèle 3.2 met en lumière l'influence de la **hauteur** de la barrière de potentiel. Sur la **figure 4**, on montre que pour deux potentiels différents, et une même énergie, la densité de probabilité est plus grande  si la barrière de potentiel est plus petite.
